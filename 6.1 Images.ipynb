```python
ageGroups = ['0-4','5-9','10-14','15-19','20-29','30-39','40-49','50-59','60-69','70-79','80+']
Counts = [957,1954,1801,1087,3003,3083,2418,2296,1639,739,462]
```

```python
import seaborn as sns
splot=sns.barplot(x=ageGroups,y=Counts)
plt.bar(ageGroups, Counts)
plt.xticks(rotation = 45)
plt.title('Age groups of COVID-19 cases \nduring the last two weeks', fontsize=16)
plt.bar_label(splot.containers[0])
```
