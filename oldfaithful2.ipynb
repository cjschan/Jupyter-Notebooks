```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
df = pd.read_csv('oldfaithful.csv')
```

```python
df
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel('Eruption time (minutes)', fontsize=14)
plt.ylabel('Waiting time to next eruption (minutes)',fontsize=14)
plt.show()
```

```python
km = KMeans(n_clusters = 2)
km = km.fit(df)
centroids = km.cluster_centers_
df['Cluster'] = km.fit_predict(df[['Eruption', 'Waiting']])
```

```python
cent0 = [i[0] for i in centroids] 
cent1 = [i[1] for i in centroids]

# add to df
df['Centroid 0'] = df.Cluster.map({0:cent0[0], 1:cent0[1]})
df['Centroid 1'] = df.Cluster.map({0:cent1[0], 1:cent1[1]})

# add colors to df
df['Colors'] = df.Cluster.map({0:'red', 1:'green'})

plt.figure(figsize=(6, 6))
plt.scatter(df.Eruption, df.Waiting, c=df.Colors, s=14)
plt.xlabel('Eruption time (minutes)', fontsize=14)
plt.ylabel('Waiting time to next eruption (minutes)',fontsize=14)
plt.show()
```

```python
df = pd.read_csv('oldfaithful.csv')

WCSS = []
k = [1,2,3,4,5]
for j in k:
    km = KMeans(n_clusters = j)
    km = km.fit(df)
    WCSS.append(km.inertia_)
```

```python
ax = plt.figure().gca()
plt.plot(k, WCSS, '*-')
plt.xlabel('Number of clusters (k)', fontsize="14")
plt.ylabel('Within-cluster sum of squares (WCSS)', fontsize="14")
```

```python
kmeans = KMeans(n_clusters=2, random_state=0, init='random', n_init=10, max_iter=10)
kmeans.fit(df)
df['Cluster']=kmeans.predict(df)
fig, ax = plt.subplots()
sns.scatterplot(data=df, x='Eruption', y='Waiting', color='blue', hue='Cluster')
ax.set(xlabel='Eruption time (minutes)')
ax.set(ylabel='Waiting time to next eruption (minutes)')
plt.title('Eruption time vs waiting time to next eruption', fontsize = 20, c='black')
plt.show()
```

```python
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], label = '1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], label = '0', edgecolors='white')
plt.xlabel('Eruption time')
plt.ylabel('Waiting time to next eruption')
plt.title('Eruption time vs. waiting time to next eruption', fontsize = 20, c='black')
plt.legend(title='Cluster')
```

```python
plt.subplot(2,1,1)
plt.scatter(df['Eruption'], df['Waiting'])
plt.title('Eruption time vs waiting time', fontsize = 20, c='black')
plt.ylabel('Waiting time', fontsize = 14)

plt.subplot(2,2,1)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], label = '1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], label = '0', edgecolors='white') 
plt.legend()
```
