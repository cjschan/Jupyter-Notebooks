```python
import pandas as pd
import seaborn as sns

titanic = pd.read_csv('https://s3-us-west-2.amazonaws.com/static-resources.zybooks.com/titanic.csv')
```

```python
df = titanic[(titanic["pclass"] == 1) & (titanic["sex"] == "male") & (titanic["age"] > 18)]
```

```python
df
```

```python
sns.stripplot(x="embark_town", y = "age", hue = "survived", data=df)
```
