```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, manifold
from sklearn.manifold import MDS, Isomap
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
sr_points, sr_color = datasets.make_swiss_roll(n_samples=500, random_state=0)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
fig.add_axes(ax)
ax.scatter(
    sr_points[:, 0], sr_points[:, 1], sr_points[:, 2], c=sr_color, s=50, alpha=0.8, cmap='viridis'
)
ax.view_init(azim=-66, elev=12)
```

```python
plt.scatter(sr_points[:, 0], sr_points[:, 2], c=sr_color, cmap='viridis')
```

```python
mds = Isomap(n_components=2)
X_transform_mds = mds.fit_transform(sr_points)
fig, plot = plt.subplots()
plt.prism()
plot.scatter(X_transform_mds[:, 0], X_transform_mds[:, 1], c=sr_color, cmap='viridis')
```

```python
mds = Isomap(n_components=1)
X_transform_mds = mds.fit_transform(sr_points)
fig, plot = plt.subplots()
plt.prism()
plot.scatter(X_transform_mds[:, 0], [0]*500, c=sr_color, cmap='viridis')
```

```python
sr_points, sr_color = datasets.make_s_curve(n_samples=500, random_state=0)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
xtick_loc = [-1,0,1]
ytick_loc = [0,1,2]
ztick_loc = [-2,-1,0,1,2]
ax.set_xticks(xtick_loc)
ax.set_yticks(ytick_loc)
ax.set_zticks(ztick_loc)
fig.add_axes(ax)
ax.scatter(
    sr_points[:, 0], sr_points[:, 1], sr_points[:, 2], c=sr_color, s=50, alpha=0.8
)
ax.view_init(azim=-66, elev=12)
```
