```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = pd.read_csv('https://s3-us-west-2.amazonaws.com/static-resources.zybooks.com/titanic.csv')
```

```python
titanic.head()
```

```python
subset1 = titanic[(titanic["pclass"] == 1) & (titanic["embarked"] == "S")]
subset2 = titanic[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)]
```

```python
sns.countplot(x="sex", color="b", data=subset1)
```

```python
# plots a vertical bar chart
sns.countplot(x="survived", color="b", data=subset2);
```
