```python
# Import all libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the abalone.csv data
abalone = pd.read_csv('abalone.csv')
```

```python
# Apply a sqrt root and logarithmic transformation to whole weight
abalone[['whole_weight_sqrt']] = np.sqrt(abalone[['whole_weight']])
abalone[['whole_weight_log']] = np.log(abalone[['whole_weight']])

# Apply a logarithmic transformation to length
abalone[['length_log']] = np.log(abalone[['length']])
```

```python
# Display the distribution of whole weight
sns.displot(abalone, x= 'whole_weight')
plt.xlabel('whole_weight', size=14)
plt.ylabel('Count', size=14)
```

```python
# Display the distribution of the square root transformation of whole weight
sns.displot(abalone, x='whole_weight_sqrt')
plt.xlabel('sqrt(whole_weight)', size=14)
plt.ylabel('Count', size=14)
```

```python
# Display a scatterplot of the logarithmic transformations of whole weight vs length
sns.scatterplot(data=abalone, x = 'whole_weight_log', y = 'length_log')
plt.xlabel('log(whole_weight)', size=14)
plt.ylabel('log(length)', size=14)
```
