```python
# Load matplotlib.pyplot and pandas
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load oldfaithfulCluster.csv data
df = pd.read_csv('oldfaithfulCluster.csv')
```

```python
plt.subplot(2,1,1)
plt.scatter(df['Eruption'], df['Waiting'])
plt.title('Eruption time vs. waiting time', fontsize = 20, c='black')
plt.ylabel('Waiting time', fontsize = 14)

plt.subplot(2,1,2)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], 
            label = '1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], 
            label = '0', edgecolors='white')
plt.xlabel('Eruption time', fontsize = 14)
plt.ylabel('Waiting time', fontsize = 14)
plt.legend()
```
