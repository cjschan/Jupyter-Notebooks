```python
import pandas as pd
import matplotlib.pyplot as plt

target = pd.read_csv('target.csv')
tgt_march = target[-19:]
tgt_volume = tgt_march[["Date","Volume"]]
tgt_close = tgt_march[["Date","Close"]]

day = int(input()) - 1
date = tgt_march.iloc[[day]].iloc[0][0]

volume_row = tgt_volume.iloc[[day]]
volume = volume_row.iloc[0][1]

close_row = tgt_close.iloc[[day]]
close = close_row.iloc[0][1]

print("The volume of TGT on " + str(date) + " is " + str(int(volume)) + ".")
print("The closing stock price of TGT on " + str(date) + " is $" + str(close) + ".")

# title
plt.title('Target (TGT) closing stock price', fontsize=20)

# x and y axis labels
plt.xlabel('Date')
plt.xticks(rotation=55)
plt.ylabel('Stock price (USD)')

# plot
plt.plot(tgt_march['Date'], tgt_march['Close'])
```
