```python
# Load matplotlib.pyplot, pandas, and numpy
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
# Load oldfaithfulCluster.csv data
df = pd.read_csv('oldfaithfulCluster.csv')
```

```python
plt.suptitle('Eruption time vs. waiting time', fontsize = 20)

plt.subplot(2,1,1)
plt.scatter(df['Eruption'], df['Waiting'])
plt.ylabel('Waiting time', fontsize = 14)

plt.subplot(2,1,2)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], 
            label = '1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], 
            label = '0', edgecolors='white')
plt.xlabel('Eruption time', fontsize = 14)
plt.ylabel('Waiting time', fontsize = 14)
plt.legend()
```

```python
plt.suptitle('Eruption time vs. waiting time')

plt.subplot(1,2,1)
plt.scatter(df['Eruption'], df['Waiting'])
plt.ylabel('Waiting time')
plt.xlabel('Eruption time')

plt.subplot(1,2,2)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], 
            label = '1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], 
            label = '0', edgecolors='white')
plt.xlabel('Eruption time')
plt.legend()
```
