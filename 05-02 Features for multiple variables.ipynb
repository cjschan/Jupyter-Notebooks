```python
# Load required packages
import pandas as pd
import seaborn as sns

# Import data from a .csv file
hawks = pd.read_csv('hawks.csv')
hawks
```

```python
# Scatterplot with Wing and Weight
p = sns.scatterplot(data=hawks, x='Wing', y='Weight')
p.set_xlabel('Wing', fontsize=14)
p.set_ylabel('Weight', fontsize=14)
```

```python
# Change point color by Species
p = sns.scatterplot(data=hawks, x='Wing', y='Weight', 
                    hue='Species')
p.set_xlabel('Wing', fontsize=14)
p.set_ylabel('Wing', fontsize=14)
```

```python
# Change point color and shape (style) by Species
p = sns.scatterplot(data=hawks, x='Wing', y='Weight', 
                    hue='Species', style='Species')
p.set_xlabel('Wing', fontsize=14)
p.set_ylabel('Wing', fontsize=14)
```

```python
# Change opacity with alpha
p = sns.scatterplot(data=hawks, x='Wing', y='Weight', 
                    hue='Species', style='Species', 
                    alpha=0.5)
p.set_xlabel('Wing', fontsize=14)
p.set_ylabel('Wing', fontsize=14)
```

```python
# Change color palette with palette
p = sns.scatterplot(data=hawks, x='Wing', y='Weight', 
                    hue='Species', style='Species', 
                    palette='Set2')
p.set_xlabel('Wing', fontsize=14)
p.set_ylabel('Wing', fontsize=14)
```
